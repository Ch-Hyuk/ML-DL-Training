## 인터엑스 프로젝트 추출 데이터 정제 작업

```python
#라이브러리 import

import pandas as pd
from openpyxl import load_workbook
import os
import re
from collections import Counter

dir_path = os.getcwd()
dir_path = os.path.join(dir_path, "file")
print(dir_path)
```

```python
# 파일 경로와 읽을 열 이름
file_name = "Project_data.xlsx"
file_path = os.path.join(dir_path,file_name)  
column_name = 'A'  

# 엑셀 파일 읽기
word_sheet = pd.read_excel(file_path, sheet_name='Sheet1', usecols=column_name)
abb_sheet = pd.read_excel(file_path, sheet_name='Sheet2', usecols=column_name)

word_sheet = word_sheet.applymap(lambda s: s.lower())
abb_sheet = abb_sheet.applymap(lambda s: s.upper())
```

```python
# 결과 엑셀 시트에 분할 저장
result_file_path = os.path.join(dir_path,'interX_Data.xlsx')
with pd.ExcelWriter(result_file_path) as writer:
    word_sheet.to_excel(writer, sheet_name='Sheet1', index=False)
    abb_sheet.to_excel(writer, sheet_name='Sheet2', index=False)
```
