## 알맞은 축약어로 정의되어 있는지 확인하는 검증 

```python
import pandas as pd
import openpyxl as op
import os

file_dir_path = os.getcwd()+'/file'
print(file_dir_path)
```

```python
def convert_string(value, conv_dict, Not_in_word_list):
    if value == None:
        return
    
    parts = value.split('_')
    convert_parts = []
    for part in parts:
        convert_part = conv_dict.get(part, '[Not in word]')
        convert_parts.append(convert_part)
        if convert_part == '[Not in word]':
            Not_in_word_list.append(part)

    print(convert_parts)
    return '_'.join(convert_parts)
```

```python
def data_modify(word_data, verify_data):
    print(word_data[['영문명', '단어 축약']])
    word_dict = dict(zip(word_data['영문명'], word_data['단어 축약']))

    NotIN_word_list = []
    verify_data['영문 축약'] = verify_data['영문명'].apply(lambda x: convert_string(x, word_dict, NotIN_word_list))
    print(verify_data['영문 축약'])
    print("정의 되어 있지 않은 단어 개수: ", len(set(NotIN_word_list)))
    print(set(NotIN_word_list))
```

```python
def data_load(file_path, file_name, ver_sheet):
    path = os.path.join(file_path, file_name)
    wb = op.load_workbook(path)

    word_sheet = list(wb['단어 정리 (세부내용 구체화)_3'].values)
    verify_sheet = list(wb[ver_sheet].values)

    w_col = word_sheet[0]
    w_data = word_sheet[1:]

    v_col = verify_sheet[0]
    v_data = verify_sheet[1:]

    word_data = pd.DataFrame(w_data, columns=w_col)
    verify_data = pd.DataFrame(v_data, columns=v_col)

    data_modify(word_data, verify_data)
```

```python
data_load(file_dir_path, '단어 및 용어 사전 정리_05.22.xlsx', '용어 정리(데이터 재정의)_7')
```
