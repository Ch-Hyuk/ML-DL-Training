```python
from konlpy.tag import Kkma

# Kkma 형태소 분석기 객체 생성
kkma = Kkma()

# 형태소 분석 결과를 저장할 함수 정의
def kkma_morphs(text):
    return kkma.morphs(text)

def kkma_li_morphs(te_list):
    return [kkma.morphs(word) for word in te_list]

def kkma_nouns(text):
    word_list = kkma.nouns(text)
    filtered_word = [word for word in word_list if not re.search(r'\d|\W', word) and len(word) > 1]
    return ''.join(filtered_word) if filtered_word else ''

def kkma_pos_t(text):
    return kkma.pos(text)

def kkma_pos(text):
    word_list = []
    morph_list = kkma.pos(text)
    for word, morp in morph_list:
        if "NN" in morp:
            word_list.append(word)
    return word_list
```

```python
from konlpy.tag import Hannanum

# Kkma 형태소 분석기 객체 생성
hannanum = Hannanum()

# 형태소 분석 결과를 저장할 함수 정의
def hannanum_morphs(text):
    return hannanum.morphs(text)

def hannanum_nouns(text):
    word_list = hannanum.nouns(text)
    filtered_word = [word for word in word_list if not re.search(r'\d|\W', word) and len(word) > 1]
    return ''.join(filtered_word) if filtered_word else ''

def hannanum_pos_t(text):
    return hannanum.pos(text)

def hannanum_pos(text):
    word_list = []
    morph_list = hannanum.pos(text)
    for word, morp in morph_list:
        if "NN" in morp:
            word_list.append(word)
    return word_list
```

```python
from konlpy.tag import Okt

# Kkma 형태소 분석기 객체 생성
t = Okt()

# 형태소 분석 결과를 저장할 함수 정의
def t_morphs(text):
    return t.morphs(text)

def t_nouns(text):
    word_list = t.nouns(text)
    filtered_word = [word for word in word_list if not re.search(r'\d|\W', word) and len(word) > 1]
    return ''.join(filtered_word) if filtered_word else ''

def t_pos_t(text):
    return t.pos(text)

def t_pos(text):
    word_list = []
    morph_list = t.pos(text)
    for word, morp in morph_list:
        if "NN" in morp:
            word_list.append(word)
    return word_list
```

```python
from konlpy.tag import Komoran

# Kkma 형태소 분석기 객체 생성
komoran = Komoran()

# 형태소 분석 결과를 저장할 함수 정의
def komoran_morphs(text):
    return komoran.morphs(text)


def komoran_nouns(text):
    word_list = komoran.nouns(text)
    filtered_word = [word for word in word_list if not re.search(r'\d|\W', word) and len(word) > 1]
    return ''.join(filtered_word) if filtered_word else ''

def komoran_pos_t(text):
    return komoran.pos(text)

def komoran_pos(text):
    word_list = []
    morph_list = komoran.pos(text)
    for word, morp in morph_list:
        if "NN" in morp:
            word_list.append(word)
    return word_list
```
