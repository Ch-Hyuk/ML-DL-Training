```python
import requests
import xml.etree.ElementTree as ET

# API 정보 설정
api_key = '6E4D94FEAF37414778AD6E695A8893C9'
base_url = 'https://krdict.korean.go.kr/api/view?key={}&q={}&req_type=json'

def get_word_definition(word):
    url = base_url.format(api_key, word)
    response = requests.get(url)
    
    #응답 본문 출력하여 확인
    print(f"URL: {url}")
    print(f"Status Code: {response.status_code}")
    print(f"Response Text: {response.text}")
    
    if response.status_code == 200:
        try:
            data = response.json()
        except ValueError:  # JSONDecodeError가 아닌 경우도 대비
            data = None
        # JSON 응답 확인
        if data and 'channel' in data and 'item' in data['channel']:
            definitions = data['channel']['item']
            return definitions
        else:
            # XML 응답으로 시도
            try:
                root = ET.fromstring(response.content)
                items = root.findall(".//item")
                definitions = [item.find("word").text for item in items]
                return definitions
            except ET.ParseError:
                return None
    return None

def can_be_verb(noun):
    possible_verb = noun + "하다"
    definitions = get_word_definition(possible_verb)
    if definitions:
        return True
    return False

# 테스트할 명사 리스트
nouns = ["검사"]


# 명사 리스트 검사
for noun in nouns:
    if can_be_verb(noun):
        print(f"{noun} -> ~하다로 동사형이 될 수 있습니다.")
    else:
        print(f"{noun} -> 동사형이 될 수 없습니다.")
```

```python
from Korpora import Korpora
from konlpy.tag import Kkma

# Korpora 로드
korpus = Korpora.load('korean_hate_speech')

# 형태소 분석기 로드
kkma = Kkma()

# 특정 명사구
target_noun_phrase = '아름다운'

# 말뭉치에서 문장 추출
sentences = korpus.get_all_texts()

# 특정 명사구가 수식어로 사용된 문장 찾기
for sentence in sentences:
    if target_noun_phrase in sentence:
        pos_tags = kkma.pos(sentence)
        
        # 수식어인지 판단
        for i, (word, pos) in enumerate(pos_tags):
            if word == target_noun_phrase.split()[0] and pos.startswith('VA'):
                if i + 1 < len(pos_tags) and pos_tags[i + 1][1].startswith('NN'):
                    print(f"문장: {sentence}")
                    print(f"'{target_noun_phrase}'가 '{pos_tags[i + 1][0]}'를 수식합니다.\n")
```
